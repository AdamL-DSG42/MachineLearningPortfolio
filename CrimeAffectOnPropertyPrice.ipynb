Reduce property price dataset to remove large outliers caused by the sale of properties such as large complexes

```python
altered_df = altered_df[altered_df["SALE_PRICE"] < 10000000]
```

Display a scatter graph of all sales and their prices per county

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, figsize = (20, 20))
ax.grid()
plt.xticks(rotation = 30)
plt.scatter(altered_df["COUNTY"], altered_df["SALE_PRICE (€1000)"], s=20)
font = {'family': 'Verdana', 
        'color':  'blue', 
        'size': 20, 
        } 
plt.xlabel('COUNTY', fontdict=font, labelpad=10)
plt.ylabel('SALE PRICE (€1000)', fontdict=font, labelpad=10)
plt.title('SALE PRICES PER COUNTY 2010 - 2021')
plt.show()
```

Display a horizontal bar chart showing the average sale price per county between 2010 and 2021

```python
horizontal_bar = altered_df.groupby('COUNTY')['SALE_PRICE (€1000)'].mean()
horizontal_bar.plot(kind="barh")

ax = horizontal_bar.plot(kind='barh', figsize=(8, 10), zorder=2, width=0.85)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")

vals = ax.get_xticks()
for tick in vals:
    ax.axvline(x=tick, linestyle='dashed', alpha=0.4, zorder=1)

ax.set_xlabel("AVERAGE SALE PRICE (€1000)", labelpad=20, weight='bold', size=12)

ax.set_ylabel("COUNTY", labelpad=20, weight='bold', size=12)

ax.set_title("AVERAGE SALE PRICE PER COUNTY 2010 - 2021")
```
