Display a scatter graph of all sales and their prices per county

This graph is used to visually show the price of every property that has been sold per county. This was used to identify any large outliers that may impact future machine learning algorithms implemented on the datasets

plt.subplots takes a number of subplots (1) and a figure size (20, 20) and returns a fig (a top level figure container for plot elements) and ax (axes of the plot)

ax.grid() configures the grid lines of the graph

plt.xticks configures the rotation angle of the labels of the x axis

plt.scatter creates a scatter plot of y vs x with size 20 marker points (s), with the format (x, y, s) below

plt.xlabel creates a label for the x axis using the text supplied, fontdict supplied and padding supplied

plt.ylabel creates a label for the y axis using the text supplied, fontdict supplied and padding supplied

plt.title sets the title of the graph

plt.show displays the graph

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, figsize = (20, 20))
ax.grid()
plt.xticks(rotation = 30)
plt.scatter(altered_df["COUNTY"], altered_df["SALE_PRICE (€1000)"], s=20)
font = {'family': 'Verdana', 
        'color':  'blue', 
        'size': 20, 
        } 
plt.xlabel('COUNTY', fontdict=font, labelpad=10)
plt.ylabel('SALE PRICE (€1000)', fontdict=font, labelpad=10)
plt.title('SALE PRICES PER COUNTY 2010 - 2021')
plt.show()
```

Display a horizontal bar chart showing the average sale price per county between 2010 and 2021

This graph was used to visually see counties where sale prices were the highest on average, allowing for these large averages to be accounted for in the case of them affecting future results

groupby groups the two selected columns from the dataframe and uses mean to get their averages

plot plots a graph of kind horizontal bar, with figure size (8, 10) and bars of width 0.85

The ax.spines code removes the border box from around the graph 

axvline adds the dashed lines on each x value to easier understand the graph

set_xlabel, set_ylabel and set_title set the x axis' label, y axis' label and graph title respectively

```python
horizontal_bar = altered_df.groupby('COUNTY')['SALE_PRICE (€1000)'].mean()

ax = horizontal_bar.plot(kind='barh', figsize=(8, 10), width=0.85)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

vals = ax.get_xticks()
for tick in vals:
    ax.axvline(x=tick, linestyle='dashed', alpha=0.4)

ax.set_xlabel("AVERAGE SALE PRICE (€1000)", labelpad=20, weight='bold', size=12)

ax.set_ylabel("COUNTY", labelpad=20, weight='bold', size=12)

ax.set_title("AVERAGE SALE PRICE PER COUNTY 2010 - 2021")
```
